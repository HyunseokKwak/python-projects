### 해양기상월보_기상자료개방포털

기상청의 기상자료 개방포털의 Open API를 이용한다

이용할 데이터는 https://data.kma.go.kr/api/selectApiList.do 의 해양기상월보이다

위 자료는 2016년 9월부터의 해양기상을 부이, 등표, 파고부이 각각의 관측값으로 제공한다

이용을 위해 활용신청이 필요하다

'https://data.kma.go.kr/OPEN_API/SEAM/2018/05/XML/note.xml' 요청을 통해 자료에 대한 정보 조회가 가능하다

buoy에서 관측한 일별기상 데이터를 받아오기위해 8번 오퍼레이션, 해양기상등표 일별기상자료를 이용하였다

```python
import requests as req
from lxml import etree

import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings(action='ignore')
```

```python
url = 'https://data.kma.go.kr/OPEN_API/SEAM/2018/05/XML/day_buoy_22101.xml'
headers = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}
```

```python
data = []
for day in days:
    tmp =[]
    chil = day.getchildren()
    for el in chil:
        tmp.append(el.text.strip())
    data.append(tmp)
```

이상의 과정을 함수로 만들자, 인자로 요청 메시지에 필요한 연도, 월, 그리고 location(station)을 받는다

location을 지정하기 위해 우선 buoy의 지점일람표 정보를 찍어보자

```python
url = 'https://data.kma.go.kr/OPEN_API/SEAM/2018/05/XML/stn_buoy.xml'
resp = req.get(url,headers = headers, verify=False) 
apiEle = etree.fromstring(resp.content)
apiTree = etree.ElementTree(apiEle)
rootNode = apiTree.getroot()
locs = rootNode.cssselect("info")
```

```python
locations =[]
for loc in locs:
    locations.append([loc.getchildren()[0].text,loc.getchildren()[1].text])
```

```python
locations
```

```python
def make_buoy_data(year,month:str,loc):
    global data # 전역변수 data를 이용한다
    url = 'https://data.kma.go.kr/OPEN_API/SEAM/{year}/{month}/XML/day_buoy_{loc}.xml'.format(**{'year':year,'month':month,'loc':loc})
    headers = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}
    #headers는 자신의 agent값을 넣어준다, 넣지 않아도 큰 문제는 없다
    
    resp = req.get(url,headers = headers, verify=False)
    apiEle = etree.fromstring(resp.content)
    apiTree = etree.ElementTree(apiEle)
    rootNode = apiTree.getroot()
    
    days = rootNode.cssselect("info")
    
    for day in days:
        tmp =[loc,year,month]
        chil = day.getchildren()
        for el in chil:
            tmp.append(el.text.strip())
        data.append(tmp)
    return
```

```python
data=[]

for loc in locations:
    for y in range(2016,2019): #16.09~ 18.05
        for m in range(1,13):
            try:
                if m<10:
                    make_buoy_data(y,'0'+str(m),loc[0])
                else:
                    make_buoy_data(y,str(m),loc[0])
            except Exception as e:
                print(e)
                continue
```
